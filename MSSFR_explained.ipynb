
In general the merger rate at any given redshift, for a specific binary is given by


\begin{equation}
Rate(Z,z) = dN_{BBH,merge} MSSFR(Z,z)/\lt mass evolved \gt
\end{equation}

In this notebook, we will mostly focus on the "Metallicity Specific Star Formation Rate":

***
\begin{equation}
MSSFR(Z,z) = SFR(z) \cdot fractions(Z,z)
\end{equation}
***


We adopt the SFR history of Madau and Dickenson:

\begin{equation}
SFR(z) = a \frac{(1 + z)^{2.7} }{[1 + (1+z)/2.9]^{b} } M_{\odot} yr^{-1} Mpc^{-3}
\end{equation}

Though we use the following for a and b, while Madau and Dickenson 2014 use the values in the brackets:

- a = 0.01 (0.015)
- b = 4.7  (5.6)
***



***
"$fractions(Z,z)$" describes "The fraction of the Universe that is some metallicity, at some given redshift".
In other words, what is the probability of finding this metallicity at some redshift

For this probability we adopt a lognormal distribution, namely


\begin{equation}
\frac{dP(z)}{dZ} = \frac{1}{Z \sigma \sqrt{2\pi}} exp\left( \frac{-[lnZ - \mu(z)]^2}{2\sigma^2} \right)
\end{equation}

usign that  
\begin{equation}
\frac{dP(z)}{dZ} = \frac{dP(z)}{dlnZ} \frac{dlnZ}{dZ} = \frac{dP(z)}{dlnZ} \frac{1}{Z}
\end{equation}

we rewrite this as: 
***
\begin{equation}
\frac{dP(z)}{dlnZ} = \frac{1}{\sigma \sqrt{2\pi}} exp\left( \frac{-[lnZ - \mu(z)]^2}{2\sigma^2} \right)
\end{equation}
***

We know that for a log-normal distribution, the mean value (for us, mean metallicity) is


\begin{equation}
\lt Z \gt = exp\left( \mu + \frac{\sigma^2}{2} \right)
\end{equation}

So we have to provide both $\mu$ and $\sigma$. We just provide $\sigma$ directly. For $\mu$ we follow Langer $\&$ Norman (2006) by approximating the mean metallicity through the Universe as: 

\begin{equation}
\lt Z \gt = Z_0 \cdot 10^{\gamma z}
\end{equation}

So we can use the last two equations to get $\mu$:

\begin{equation}
\mu = ln( Z_0 \cdot 10^{\gamma z} ) - \frac{\sigma^2}{2}
\end{equation}

***

To get the fraction of the Universe with some metallicity Z we should compute:

\begin{equation}
fraction (Z,z) = \frac{ \int^{Z2}_{Z1} \frac{dP(z)}{dlnZ} }{   \int^{Zmax}_{Zmin} \frac{dP(z)}{dlnZ}  }
\end{equation}

To make sure our full probability is normalized, we thus compute the denominator:

\begin{equation}
C_{norm} =  \int^{ln(Zmax) = 0.0}_{ln(Zmin) =-12} \frac{dP(z)}{dlnZ}  
\end{equation}


Prob of drawing in COMPAS

\begin{equation}
P_{draw, COMPAS} = \frac{ \int^{Z2}_{Z1} ln(Z)^0 }{   \int^{Zmax, COMPAS}_{Zmin, COMPAS} ln(Z)^0   } 
= \frac{ln(Z_2) - ln(Z_1)}{ln(Z_{max, COMPAS}) - ln(Z_{min, COMPAS})}
\end{equation}

\begin{equation}
P_{draw, COMPAS} = \frac{1}{ln(Z_{max, COMPAS}) - ln(Z_{min, COMPAS})}
\end{equation}


\begin{equation}
MSSFR(Z,z) = SFR(z) \cdot \frac{fractions(Z,z)}{P_{draw, COMPAS}}
\end{equation}



```python
def get_metallicity_distribution(min_logZ  =-12.0, max_logZ  =0.0, step_logZ = 0.01,
                                 min_logZ_COMPAS = np.log(1e-4),max_logZ_COMPAS = np.log(0.03),
                                 max_redshift = 10.0,redshift_step = 0.01,
                                 Z0=0.035, alpha=-0.23, sigma=0.39):
    """
    Calculate the distribution of metallicities at different redshifts using Neijssel+19 Eq.7-9
    (which is based on Madau & Dickinson 2014)

    NOTE: This assumes that metallicities in COMPAS are drawn from a flat in log distribution

    Args:
        redshifts          --> [list of floats] List of redshifts at which to calculate things
        min_logZ_COMPAS    --> [float]          Minimum logZ value that COMPAS samples
        max_logZ_COMPAS    --> [float]          Maximum logZ value that COMPAS samples
        Z0                 --> [float]          Parameter used for calculating metallicity density dist
        alpha              --> [float]          Parameter used for calculating metallicity density dist
        sigma              --> [float]          Parameter used for calculating metallicity density dist
        min_logZ           --> [float]          Minimum logZ at which to calculate dPdlogZ
        max_logZ           --> [float]          Maximum logZ at which to calculate dPdlogZ
        step_logZ          --> [float]          Size of logZ steps to take in finding a Z range

    Returns:
        dPdlogZ            --> [2D float array] Probability of getting a particular logZ at a certain redshift
        metallicities      --> [list of floats] Metallicities at which dPdlogZ is evaluated
        p_draw_metallicity --> float            Probability of drawing a certain metallicity in COMPAS (float because assuming uniform)
    """
    ##################################
    redshifts = np.arange(0, max_redshift + redshift_step, redshift_step)
    print('redshifts', redshifts, 'N redshifts', len(redshifts))

    # use Neijssel+19 Eq. 9 to find means
    mean_metallicities = Z0 * 10**(alpha * redshifts)
    mu_metallicities = np.log(mean_metallicities) - sigma**2/2

    # create a range of metallicities
    log_metallicities = np.arange(min_logZ, max_logZ + step_logZ, step_logZ)
    metallicities = np.exp(log_metallicities)
    print('metallicities', metallicities, 'N metallicities', len(metallicities))

    # use Neijessel+19 Eq. 7 to find probabilities (without the factor of 1/Z since this is dp/dlogZ not dp/dZ)
    dPdlogZ = 1 / (sigma * np.sqrt(2*np.pi)) * np.exp(-(log_metallicities - mu_metallicities[:,np.newaxis])**2 / (2 * sigma**2))

    # normalise the distribution
    norm = dPdlogZ.sum(axis=1) * step_logZ
    dPdlogZ = dPdlogZ / norm[:,np.newaxis]


    #Now we are not really using some interval dZ, rather, we take a singular value of some metallicity per binary. 
    #We have to normalize to account for the sampled range of metalliciities. We sample metallicies in COMPAS uniformly in log

    # assume a flat in log distribution in metallicity to find probability of drawing Z
    p_draw_metallicity = 1 / (max_logZ_COMPAS - min_logZ_COMPAS)

    print(len(dPdlogZ[:, 0]), metallicities[0] )
    
    return dPdlogZ, redshifts, metallicities, p_draw_metallicity

```
